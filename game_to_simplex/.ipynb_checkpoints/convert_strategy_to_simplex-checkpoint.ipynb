```python
from LPP import *
from LPPSolver import *
import numpy as np
```

```python
def convert_to_lpp(str_matr, player):
    if player == 'A':
        A = -str_matr.T
        b = -np.ones(str_matr.shape[1])
        c = -np.ones(str_matr.shape[0])
        return LinearProgrammingProblem(A, b, c, alternative=True)
    elif player == 'B':
        A = str_matr
        b = np.ones(str_matr.shape[0])
        c = np.ones(str_matr.shape[1])
        return LinearProgrammingProblem(A, b, c)
```

```python
stratigy_matrix = np.array([
    [6, 17, 16, 18, 15],
    [18, 8, 16, 8, 8],
    [6, 13, 18, 4, 3],
    [15, 14, 2, 18, 19]
])
```

# Simplex for A player

```python
problem = convert_to_lpp(stratigy_matrix, 'A')
solver = LPPSolver(problem)
solver.solve()
```

```python
print(solver.get_optimal_vector())
```

# Simplex for player B

```python
problem = convert_to_lpp(stratigy_matrix, 'B')
solver = LPPSolver(problem)
solver.solve()
```

```python
print(solver.get_optimal_vector())
```
